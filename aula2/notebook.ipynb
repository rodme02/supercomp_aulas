```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
tempos_limitado_15 = [0.866288, 0.863618, 0.923806, 0.855353, 0.858573]
tempos_limitado_30 = [5.84632, 2.64763, 1.78501, 3.41303, 1.71383]
tempos_limitado_50 = [3.76861, 2.87334, 6.2232, 3.58457, 2.8723]
tempos_limitado_100 = [26.454, 9.28777, 11.4164, 12.5439, 10.686]

tempos_buffado_15 = [0.234187, 0.234118, 0.233292, 0.233605, 0.233117]
tempos_buffado_30 = [0.466027, 0.462359, 0.471431, 0.463495, 0.463863]
tempos_buffado_50 = [0.768637, 0.767108, 0.767164, 0.769888, 0.767402]
tempos_buffado_100 = [1.53231, 1.53393, 1.52937, 1.53077, 1.5343]
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Tempo de execução para slurm limitado e buffado')
plt.ylabel('Tempo (s)')
plt.xlabel('Número de iterações')
plt.plot([15, 30, 50, 100], [np.mean(tempos_limitado_15), np.mean(tempos_limitado_30), np.mean(tempos_limitado_50), np.mean(tempos_limitado_100)], label='Limitado')
plt.plot([15, 30, 50, 100], [np.mean(tempos_buffado_15), np.mean(tempos_buffado_30), np.mean(tempos_buffado_50), np.mean(tempos_buffado_100)], label='Buffado')
plt.legend()
```

# Conclusão

Percebemos que a configuração de recursos é muito importante. O tempo de execução para uma alocação limitada de recursos aumenta significativamente com o número de iteraçõe, enquando a alocação de recursos buffada não sofre tanto com o aumento do número de iterações.

Porém, é necessário ter um equilíbro na alocação de recursos, se alocarmos mais recursos do que o necessário a task pode ficar muito tempo na fila. Por isso, é importante fazer testes para encontrar a quantidade ideal de recursos a serem alocados, para que a task não fique muito tempo na fila e nem demore muito para ser executada.
