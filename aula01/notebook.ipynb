```python
import matplotlib.pyplot as plt
import numpy as np
```

## Tamanhos das matrizes e tempos de execução

```python
tamanhos_matriz = [200, 300, 400, 500, 600, 700, 800, 900, 1000, 1100, 1200, 1300]

tempos_python = [0.82, 2.43, 5.81, 11.25, 23.78, 31.72, 47.46, 66.76, 111.07, 126.72, 164.95, 222.50]
tempos_cpp = [0.044, 0.146, 0.35, 0.685, 1.185, 1.863, 2.838, 4.042, 5.52, 7.492, 9.88, 12.306]
tempos_openmp = [0.014, 0.043, 0.099, 0.197, 0.352, 0.722, 1.093, 1.589, 1.622, 2.892, 2.74, 4.987]
```

## Gráficos

```python
plt.plot(tamanhos_matriz, tempos_python, label='Python')
plt.plot(tamanhos_matriz, tempos_cpp, label='C++')
plt.plot(tamanhos_matriz, tempos_openmp, label='C++ com paralelismo')

plt.xlabel('Tamanho da matriz')
plt.ylabel('Tempo médio (s)')
plt.title('Tempo médio de execução')
plt.legend()
plt.show()
```

```python
plt.plot(tamanhos_matriz, tempos_cpp, label='C++')
plt.plot(tamanhos_matriz, tempos_openmp, label='C++ com paralelismo')

plt.xlabel('Tamanho da matriz')
plt.ylabel('Tempo médio (s)')
plt.title('Tempo médio de execução C++ com e sem paralelismo')
plt.legend()
plt.show()
```

## Conclusões

### Discuta o impacto do ambiente HPC no desempenho das operações e como o paralelismo pode escalar em um cluster de múltiplos nós.


O ambiente HPC melhora o desempenho das operações através da execução paralela em múltiplos núcleos e processadores, reduzindo assim os tempos de processamento. A escalabilidade é alcançada distribuindo tarefas entre vários nós de um cluster, permitindo aumentos quase lineares no desempenho à medida que mais nós são adicionados.
