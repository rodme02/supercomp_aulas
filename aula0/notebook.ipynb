```python
import matplotlib.pyplot as plt
import numpy as np
```

## Tamanhos das matrizes e tempos de execução

```python
tamanhos_matriz = [200, 300, 900, 1300]

tempos_python_200 = [0.68, 0.73, 0.69]
tempos_python_300 = [2.61, 2.62, 2.65]
tempos_python_900 = [78.32, 82.49, 89.51]
tempos_python_1300 = [337.54, 314.43, 314.04]

tempos_cpp_200 = [0.065, 0.064, 0.069]
tempos_cpp_300 = [0.25, 0.29, 0.25]
tempos_cpp_900 = [6.78, 6.74, 6.68]
tempos_cpp_1300 = [31.83, 36.87, 34.03]

tempos_openmp_200 = [0.065, 0.063, 0.064]
tempos_openmp_300 = [0.22, 0.23, 0.26]
tempos_openmp_900 = [7.37, 8.91, 8.77]
tempos_openmp_1300 = [36.29, 36.75, 36.97]

# Médias
media_python_200 = np.mean(tempos_python_200)
media_python_300 = np.mean(tempos_python_300)
media_python_900 = np.mean(tempos_python_900)
media_python_1300 = np.mean(tempos_python_1300)
medias_python = [media_python_200, media_python_300, media_python_900, media_python_1300]

media_cpp_200 = np.mean(tempos_cpp_200)
media_cpp_300 = np.mean(tempos_cpp_300)
media_cpp_900 = np.mean(tempos_cpp_900)
media_cpp_1300 = np.mean(tempos_cpp_1300)
medias_cpp = [media_cpp_200, media_cpp_300, media_cpp_900, media_cpp_1300]

media_openmp_200 = np.mean(tempos_openmp_200)
media_openmp_300 = np.mean(tempos_openmp_300)
media_openmp_900 = np.mean(tempos_openmp_900)
media_openmp_1300 = np.mean(tempos_openmp_1300)
medias_openmp = [media_openmp_200, media_openmp_300, media_openmp_900, media_openmp_1300]
```

## Gráficos

```python
plt.plot(tamanhos_matriz, medias_python, label='Python')
plt.plot(tamanhos_matriz, medias_cpp, label='C++')
plt.plot(tamanhos_matriz, medias_openmp, label='C++ com paralelismo')

plt.xlabel('Tamanho da matriz')
plt.ylabel('Tempo médio (s)')
plt.title('Tempo médio de execução')
plt.legend()
plt.show()
```

```python
plt.plot(tamanhos_matriz, medias_cpp, label='C++')
plt.plot(tamanhos_matriz, medias_openmp, label='C++ com paralelismo')

plt.xlabel('Tamanho da matriz')
plt.ylabel('Tempo médio (s)')
plt.title('Tempo médio de execução C++ com e sem paralelismo')
plt.legend()
plt.show()
```

## Conclusões

### Faça uma análise comparativa sobre o impacto do paralelismo no desempenho de acordo com a complexidade do problema.

Conforme o tamanho das matrizes aumenta, o paralelismo ajuda a reduzir o tempo de execução. Isso é mais evidente para matrizes maiores, onde a diferença entre os tempos de execução é mais significativa. Para matrizes menores, o paralelismo não traz tanto benefício.

### Comente sobre como você acha que este problema pode ser abordado em um ambiente de HPC.

Em ambientes de HPC, a multiplicação de matrizes muito grandes pode ser otimizada utilizando arquiteturas paralelas avançadas. Isso inclui o emprego de múltiplos nós de processamento, cada um com CPUs ou GPUs multi-core, para dividir e conquistar as tarefas computacionais intensivas. O HPC permite uma implementação eficiente de paralelismo em grande escala, reduzindo significativamente o tempo de execução através da distribuição eficaz do trabalho e minimizando o overhead de comunicação entre os processos.
